```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina' #make plots look better
```

```python
df_cars = pd.read_csv("data/cars.csv")
df_cars.head()
```

```python
#miles per gallon for each car with same number of cylinders
avg_mpg = df_cars.groupby('CYL').mean()['MPG'] 
avg_mpg
```

```python
avg_wgt = df_cars.groupby('CYL').mean()['WGT']
#average weight for each car with the same cylinder
```

```python
#average miles per gallon for each car with the same weight
avg_mpg_per_weight = df_cars.groupby('WGT').mean()['MPG']
```

```python
#Get the unique list of cylinders in numerical order
cyl = sorted(df_cars['CYL'].unique())

#Get a list of all mpg values for three specific cylinder sizes 
cyl4 = df_cars[df_cars['CYL'] == 4]['MPG'].values
cyl6 = df_cars[df_cars['CYL'] == 6]['MPG'].values
cyl8 = df_cars[df_cars['CYL'] == 8]['MPG'].values
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize = (3,3))

ax.hist(df_cars['WGT'], bins = 50)

plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (3,3))

#plots below give the same plot with same x-axis and y-axis
ax.plot(cyl, avg_mpg)
ax.plot(avg_mpg) #plot function looks at the series and plots indices in x-axis
ax.set_xlabel("Cylinders")
ax.set_ylabel("Avg mpg")
ax.set_title("Effect of number of cylinders on MPG")

plt.show()
```

## Heatmap

```python
C = df_cars.corr().abs()
```

```python
fig, ax = plt.subplots(1,1, figsize = (3,3))

ax.imshow(C, cmap = "Blues")

ax.set_xticks(range(4))
ax.set_xticklabels(C.columns)
ax.set_yticks(range(4))
ax.set_yticklabels(C.columns)

ax.text(0, 1, f"{C.iloc[0, 1]:.2f}", horizontalalignment ='center')
ax.text(1, 0, f"{C.iloc[1, 0]:.2f}", horizontalalignment = 'center')

plt.savefig("heat.png", bbox_inches = "tight", pad_inches = 10)
```
