```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches #for drawing shapes
%config InlineBackend.figure_format = 'retina'
```

```python
data = [96.2, 86.4, 72.0, 19.8, 34.7, 63.2]
names = ['', 'Bahrain', 'Bangladesh', 'Barbados', 'Belarus', 'Belgium', 'Belize']
```

```python
for i in np.arange(len(data)): 
    print(data[i])
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (4, 3))

#set x and y limits
ax.set_ylim(0.1, 1.3, 0.2)
ax.set_xlim(0, 100)

#set tick labels
ax.set_yticklabels(names)
ax.set_xticklabels([])

#label graph
ax.text(20, 1.4, "Country Statistics", horizontalalignment = 'left', verticalalignment = 'center', 
       fontsize = 15)

ax.text(0, .03, "0", horizontalalignment = 'center', verticalalignment = 'center')

#add each of the plots and text above the plot
for i in np.arange(len(data)): 
    ax.plot([0, data[i]], [.2 * (i + 1),.2 * (i + 1)] , "-", lw = 1, c = "black")
    ax.plot(data[i], .2 * (i + 1), ">", c = "black", markersize = 3)
    ax.text(data[i], .2 * (i + 1) + .05, data[i], fontsize = 8, horizontalalignment = 'center', 
           verticalalignment = 'center')

#Get rid of margins and lines within the graph
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(.5)
ax.spines['bottom'].set_visible(False)

#add gridlines with low linewidth
ax.grid(axis = 'y', linewidth = .5)

#save image as png
#plt.savefig("heat.png", bbox_inches = "tight", pad_inches = 10)

#get rid of ticks on x and y axis
plt.tick_params(axis='y',which='both', bottom=False, top=False, labelbottom=False)
plt.tick_params(axis='x',which='both', bottom=False, top=False, labelbottom=False)
plt.show()
```
