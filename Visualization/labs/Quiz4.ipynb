```python
df_bull = pd.read_csv('data/Train.csv', parse_dates = ['saledate'])
df_bull.head()
```

```python
x = pd.DataFrame(df_bull.corr().unstack()).reset_index().rename(columns= {0: "corr"})
x['corr'] = x['corr'].apply(lambda x: -x if x < 0 else x)
x = x[x['corr'] < 1]
x = x[x['level_0'] == 'SalePrice']
#np.round(x.sort_values(by = "corr", ascending = False).iloc[0]['corr'], 3)
x
```
